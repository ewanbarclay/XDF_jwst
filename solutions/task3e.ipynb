# Task 3e Solution

```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
filters = [ 'f090w', 'f150w', 'f200w','f277w','f356w','f444w']
```

```python
iso_cat = pickle.load(open('data/cat.p', 'rb'))  # read in signal catalogue
aper_cat = pickle.load(open('data/aper_cat.p', 'rb'))  # read in signal catalogue for aperture
```

```python
iso_fluxes = {f: iso_cat[f'{f}_signal'] for f in filters} # convert to flux/nJy
aper_fluxes = {f: aper_cat[f'{f}_signal'] for f in filters} # convert to flux/nJy
```

```python
iso_r1 = iso_fluxes['f150w']/iso_fluxes['f200w'] # break colour (mag) - usually on y-axis
iso_r2 = iso_fluxes['f090w']/iso_fluxes['f150w'] # slope colour (mag) - usually on x-axis
```

```python
aper_r1 = aper_fluxes['f150w']/aper_fluxes['f200w'] # break colour (mag) - usually on y-axis
aper_r2 = aper_fluxes['f090w']/aper_fluxes['f150w'] # slope colour (mag) - usually on x-axis
```

--- plot all isophotal flux ratios for all objects

```python
plt.scatter(iso_r1, iso_r2, s=5, c='k', marker = '^')
# --- plot all aperture flux ratios for all objects
plt.scatter(aper_r1, aper_r2, s=5, c='r', marker = 'v')

# Connect each pair
for ir1, ir2, ar1, ar2 in zip(iso_r1,iso_r2,aper_r1,aper_r2):
    plt.plot([ir1, ar1], [ir2, ar2], c='k', alpha=0.3, lw=1)
plt.xlim([-0.3,2])
plt.ylim([-0.3,2])
plt.show()
```
